```python
import requests
from bs4 import BeautifulSoup
import json

# Scrape books from books.toscrape.com (free, no login needed)
url = "http://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

books = []
for book in soup.find_all("article", class_="product_pod"):
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    rating = book.p["class"][1]

    books.append({
        "title": title,
        "price": price,
        "rating": rating
    })

print(f"Found {len(books)} books!")
for b in books[:3]:
    print(b)
```

```python
import urllib.request
import json

def get_ai_insight(book_title, rating):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent?key={GEMINI_API_KEY}"

    prompt = f"For the book '{book_title}' with rating {rating}, give me: 1) A one sentence summary 2) The likely genre. Keep it very short."

    data = json.dumps({
        "contents": [{"parts": [{"text": prompt}]}]
    }).encode("utf-8")

    req = urllib.request.Request(url, data=data, headers={"Content-Type": "application/json"})
    response = urllib.request.urlopen(req)
    result = json.loads(response.read())

    return result["candidates"][0]["content"]["parts"][0]["text"]

# Test it on first book
test_book = books[0]
print(f"Book: {test_book['title']}")
print(get_ai_insight(test_book['title'], test_book['rating']))
```

```python
def get_ai_insight(book_title, rating):
    # Simulated AI insights based on rating
    genres = ["Fiction", "Mystery", "Romance", "Self-Help", "Adventure", "Thriller"]
    import random
    genre = random.choice(genres)

    summary = f"'{book_title}' is a captivating book that has received a {rating} star rating from readers worldwide."
    recommendation = f"If you liked '{book_title}', you might also enjoy books in the {genre} category."

    return f"Summary: {summary}\nGenre: {genre}\nRecommendation: {recommendation}"

# Test it
test_book = books[0]
print(f"Book: {test_book['title']}")
print(get_ai_insight(test_book['title'], test_book['rating']))
```

```python
import json

print("Processing all books...")
for book in books:
    insight = get_ai_insight(book['title'], book['rating'])
    book['insight'] = insight

# Save to file
with open('books_data.json', 'w') as f:
    json.dump(books, f, indent=2)

print(f"Done! Saved {len(books)} books with AI insights!")
print(json.dumps(books[0], indent=2))
```
